```python
# Random Module

import random

# Generate Random Integers

random_integer = random.randint(1, 10)    # [1 , 10]
print(random_integer)

# Generating Random floating point numbers

random_float = random.random()            # [0.0 , 1.0)
print(random_float)

# Random decimal numbers between 0 and more than 1

print(random_float * 5)                   # [0.0 , 5.0)

love_score = random.randint(0,100)
print(f"Your love score is {love_score}.")
```

```python
# Offset and Appending Items to Lists

fruits = ["Cherry", "Apple", "Pear" , "Banana"]

print(fruits)
print(fruits[0])   # computers counts from 0

# Count items from end of list

print(fruits[-1])      

# Changing item from list

fruits[1] = "Pineapple"   

print(fruits)

#  Adding new item to end of the list

fruits.append("Berry")

print(fruits)

# Adding new items to end of the list

fruits.extend(["Strawberry" , "Blueberry"])

print(fruits)
```

```python
# Project: Rock Paper Scissors

# Make a rock, paper, scissors game.
    # Rock wins against scissors.
    # Scissors win against paper.
    # Paper wins against rock.
# Start the game by asking the player:
    # "What do you choose? Type 0 for Rock, 1 for Paper or 2 for Scissors."

import random

rock = '''
    _______
---'   ____)
      (_____)   
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

images = [rock, paper, scissors]

player = int(input("What do you choose? Type 0 for Rock, 1 for Paper or 2 for Scissors.\n"))

if player >= 3 or player < 0: 
    print("You typed an invalid number, you lose!") 
else:
    print(images[player])

    cpu = random.randint(0, 2)
    print("Computer chose:")
    print(images[cpu])

    if player == 0 and cpu == 2:
        print("You win!")
    elif player == 2 and cpu == 0:
        print("You lose")
    elif player > cpu:
        print("You win!")
    elif player < cpu:
        print("You lose")
    elif cpu == player:
        print("It's a draw")
```
